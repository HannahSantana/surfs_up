```python
%matplotlib notebook
from matplotlib import style
style.use('fivethirtyeight')
import matplotlib.pyplot as plt
%matplotlib inline
import pandas as pd
```

```python
import sqlalchemy
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
from sqlalchemy import create_engine, func
%ls
```

```python
engine = create_engine("sqlite:///hawaii_new.sqlite")
```

```python
Base = automap_base()
Base.prepare(engine, reflect=True)
```

```python
Base.classes.items()
```

```python
Base.classes.keys()
```

```python
Measurement = Base.classes.measurement
Station = Base.classes.station
```

```python
session = Session(engine)
```

## Precipitation Analysis 

```python
precipitation= session.query(Measurement.date,
                        Measurement.prcp)\
            .filter(Measurement.date>2017)\
            .order_by(Measurement.date.desc())
precipitation
```

```python
for row in precipitation:
    print(row)
```

```python
dict_values = {d:t for d,t in precipitation}
```

```python
df = pd.read_sql_query(precipitation.statement,engine, index_col='date')
```

```python
df.groupby('date').sum()
df.rename(columns = {'prcp': "Precipitation"}, inplace= True)
df.dropna(axis = 0, how = 'any', inplace = True)
df.head()
```

```python
df.plot(kind = 'line',figsize=(20,10))
plt.title('Precipitation')
plt.ylabel('Rainfall in mm')
plt.xlabel('date')
```

```python
sum_statistics = df.describe()
sum_statistics
```

```python
def dates(start_date, end_date):

    tobs_date = session.query(
        func.min(Measurement.tobs.label('min_temp')),
        func.max(Measurement.tobs.label('max_temp')),
        func.avg(Measurement.tobs.label('avg_temp')),
        Measurement.date)\
    .filter(Measurement.date.between('{}'.format(start_date),'{}'.format(end_date)))
    return {"data": [x for x in tobs_date][0]}

x = dates('2017-01-01', '2017-05-01')
x
```

## Station Analysis

```python
Station.classes.items()
```

```python
station = session.query(Station.station,
                        Station.name,
                        Station.latitude,
                        Station.longitude,
                        Station.elevation)

station
```

```python
for row in station:
    print(row)
```

```python
stations = []
for row in session.query(Station):
    stations.append(
        {'id':row.station,
        'name': row.name}
    )
stations
```

```python
df1 = pd.read_sql_query(station.statement,engine)
num_stations = len(df1)
num_stations
```

```python
station_analysis = session.query(Station.station,
                        Station.name,
                        Station.latitude,
                        Station.longitude,
                        Station.elevation,
                        Measurement.tobs,
                        (func.count(Measurement.prcp)))\
    .filter(Station.station== Measurement.station)\
    .group_by(Station.station)\
    .order_by(func.count(Measurement.prcp).desc())
```

```python
for row in station_analysis:
    print(row)
```

```python
df2 = pd.read_sql_query(station_analysis.statement, engine)
df2.rename(columns = {'count_1': 'Prcp Observations'})
```

```python
tobs = session.query(Station.station,
                        Station.name,
                        Station.latitude,
                        Station.longitude,
                        Station.elevation,
                         Measurement.prcp,
                         Measurement.tobs,
                        (func.count(Measurement.tobs)))\
    .filter(Station.station== Measurement.station)\
    .group_by(Station.station)\
    .order_by(func.count(Measurement.tobs).desc())
```

```python
tobs_df = pd.read_sql_query(tobs.statement, engine)
tobs_df
```

## Temperature Analysis Histogram - Grouped by Station

```python
tobs_df.hist(column='tobs', bins= 12, rwidth=1, align = "left",figsize=(20,10))
plt.title('Temperature Analysis - Grouped by Station')
plt.ylabel('Frequency')
plt.xlabel('Observed Temperature')
plt.legend('tobs', loc = 'upper left')
```
